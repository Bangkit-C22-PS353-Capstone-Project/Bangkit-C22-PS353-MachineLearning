```python
# import that smooth library

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
# prepare the dataset, 
# download data from google drive
# https://drive.google.com/file/d/1JSKQ9LJgXFz1jlELSrKO-XQEMr179Ms8/view?usp=sharing

!gdown --id 1JSKQ9LJgXFz1jlELSrKO-XQEMr179Ms8
```

```python
# read the dataset and convert into csv
dataset = pd.read_csv('Beras Kualitas Bawah II (kg) 2021-2022.csv')
# show me that csv
print(dataset)
```

```python
# convert the colum Date from object type to date type
dataset['Date'] =  pd.to_datetime(dataset['Date'], format='%d/%m/%Y')
```

```python
# see the result
print(dataset)
```

```python
# checking the datatypes
print(dataset.dtypes)
```

```python
# put x and y axis respectively
fig = px.line(dataset, x='Date', y='Price')

# this code is to add the range slider so we can see better
fig.update_xaxes(rangeslider_visible=True)

# for showing the plot
fig.show()
```
